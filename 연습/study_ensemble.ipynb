#### 앙상블이란?
여러가지 머신러닝 클래스들을 이용해서 예측하는것
- classifier -> 다수결의 원칙으로 결정
- regressor -> 평균값으로 결정
##### => 이러한 것들을 'voting'을 based 했다고 말함, 결과적으로 취합(aggregate)의 방법론 사용

머신러닝 클래스의 갯수는 유한하기 때문에 한계가 있기 때문에 bagging을 사용 

#### Bagging
- 1개의 알고리즘을 이용하지만 그것을 많이 쓰는 것 
- 즉, 데이터를 변형 시켜서 활용
- 병렬 처리 가능

배깅을 사용한 Decision tree의 앙상블 모델은 랜덤 포레스트

#### Boosting
- 취합을 하지 않고 점진적으로 학습시켜서 연결

1. 앞의 모델에서 틀린 데이터에 대해 가중치를 높여가며 stronger learner를 만드는 AdaBoost
2. 앞의 모델이 내놓은 정답의 오차에서 오차를 학습시키며 stronger learner를 만드는 Gradient Boosting 



### Voting

```python
from sklearn.ensemble import VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import os
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
X = pd.DataFrame(cancer.data)
y = pd.Series(cancer.target)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X.values,y.values,random_state=1,stratify=y.values)
print(y.value_counts())
```

```python
lr_model = LogisticRegression(solver='lbfgs')
knn_model = KNeighborsClassifier(n_neighbors=3)
svc_model=LinearSVC()
dt_model = DecisionTreeClassifier()
```

```python
ensemble=VotingClassifier(
estimators=[
    ('lr',lr_model),
    ('knn',knn_model),
    ('svc',svc_model),
    ('dt',dt_model)
])
```

```python
lr_model.fit(X_train,y_train)
knn_model.fit(X_train,y_train)
svc_model.fit(X_train,y_train)
dt_model.fit(X_train,y_train)
ensemble.fit(X_train,y_train)
```

```python
print('LR:',lr_model.score(X_train,y_train))
print('knn:',knn_model.score(X_train,y_train))
print('svc:',svc_model.score(X_train,y_train))
print('dt:',dt_model.score(X_train,y_train))
print('Ensemble:',ensemble.score(X_train,y_train))
```

### Bagging

```python
from sklearn.ensemble import BaggingClassifier
ensemble1000 = BaggingClassifier(base_estimator=DecisionTreeClassifier(),
                                n_estimators=1000,
                                n_jobs=-1,
                                max_samples=0.5,
                                max_features=0.5,
                                random_state=1).fit(X_train,y_train)
```

```python
print('dt:',dt_model.score(X_train,y_train))
print('ensemble1000',ensemble1000.score(X_train,y_train))
```
